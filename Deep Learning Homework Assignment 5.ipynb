```python
from sklearn.datasets import fetch_mldata
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib
import matplotlib.pyplot as plt

mnist = fetch_mldata('MNIST original')
np.random.seed(0)
m = 2000
idx = np.random.permutation(60000)[:m]
print ('The size of the traning set is: ', mnist["data"][idx].shape)
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=1)
rnd_clf.fit(mnist["data"][idx], mnist["target"][idx])
for i,score in zip(range(784), rnd_clf.feature_importances_):
    print(i, score)
```

```python
new_score = rnd_clf.feature_importances_.reshape(28,28)
plt.imshow(new_score, matplotlib.cm.hot)
cbar = plt.colorbar(ticks=[rnd_clf.feature_importances_.min(), rnd_clf.feature_importances_.max()])
cbar.ax.set_yticklabels(['Not important', 'Very important'])
plt.show()
```
